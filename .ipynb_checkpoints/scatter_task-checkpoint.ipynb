```python
import pandas as pd
import matplotlib.pyplot as plt
df  = pd.read_excel("trend.xlsx")
%matplotlib inline

x = df['XY']
y = df ['Z']
classes = df.iloc[:, 2].values

markers = ['*','o','^','+','s','d','v','>','<','h','x']
colors = ['r','g','b','yellow','black','cyan','magenta','purple','orange','Tan','Pink']

info = []
for i in range(len(markers)):
    info.append([markers[i], colors[i]])
#print(info)

keys = list(set(classes))
values = info
#print(len(keys), len(values))

data = dict(zip(keys, values))

fig = plt.figure()

plt.title('Build Capacity')

for i in range(len(classes)):
    class_name = classes[i]
    plt.scatter(x[i], y[i], c=data[class_name][1], marker = data[class_name][0])

plt.xlabel('Area (X*Y-mm^2)')
plt.ylabel('Height (Z-mm)')
plt.legend(set(classes), loc='center right', bbox_to_anchor=(1.28, 0.5))

fig.savefig('scatter_plot.png',bbox_inches='tight')
```
